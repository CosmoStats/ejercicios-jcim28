```python
from getdist import loadMCSamples, plots

# Carga cada corrida (usa el prefijo de output que le diste a cada caso)
samples_bao      = loadMCSamples("./test_multi_sn")
samples_sn       = loadMCSamples("./bao_full")
samples_bao_sn   = loadMCSamples("./bao_sn")
samples_planck   = loadMCSamples("./plank")
samples_full  = loadMCSamples("./full")

# Nombres legibles para la leyenda (si no, usa el nombre del root)
samples_bao.label     = "BAO"
samples_sn.label      = "Pantheon (SN)"
samples_bao_sn.label  = "BAO + SN"
samples_planck.label  = "Planck"
samples_full.label = "Planck + BAO + SN"

g = plots.get_subplot_plotter()
g.triangle_plot(
    [samples_bao, samples_sn, samples_bao_sn, samples_planck, samples_full],
    ['Omega_sfdm_1', 'm_sfdm1','omega_b', 'omega_cdm', 'H0'],   # tus params CDM, por ejemplo
    filled=True,
    legend_labels=['BAO', 'Pantheon (SN)', 'BAO + SN', 'Planck', 'Planck + BAO + SN']
)
```

```python
g.export('triangle_plot0.png')
```

```python
g = plots.get_subplot_plotter()
g.triangle_plot(
    [samples_planck, samples_full],
    ['Omega_sfdm_1', 'm_sfdm1','omega_b', 'omega_cdm', 'H0'],   # tus params CDM, por ejemplo
    filled=True,
    legend_labels=['Planck', 'Planck + BAO + SN']
)
```

```python
g.export('triangle_plot.png')
```

```python
param_labels = {
    'Omega_sfdm_1': r'\Omega_{\rm sfdm}',
    'm_sfdm1':      r'\log_{10}(m_{\rm sfdm}/{\rm eV})',
    'omega_b':      r'\Omega_b h^2',
    'omega_cdm':    r'\Omega_{\rm cdm} h^2',
    'H0':           r'H_0',
}

all_samples = [samples_bao, samples_sn, samples_bao_sn, samples_planck, samples_full]

for s in all_samples:
    names = s.getParamNames()
    for name, label in param_labels.items():
        par = names.parWithName(name)
        if par is not None:
            par.label = label
```

```python
g = plots.get_subplot_plotter()
g.triangle_plot(
    all_samples,
    ['Omega_sfdm_1', 'm_sfdm1', 'omega_b', 'omega_cdm', 'H0'],
    filled=True,
    legend_labels=['BAO', 'Pantheon (SN)', 'BAO + SN', 'Planck', 'Planck + BAO + SN']
)
```

```python
g.export('triangle_plot1.png')
```
