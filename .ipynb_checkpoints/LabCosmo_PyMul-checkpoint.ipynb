# PyMultiNest: ajustes de diferentes cosmologías con datos de Pantheon+

#### Cargamos paquetes necesarios

```python
import json
import numpy as np
import pymultinest
import matplotlib.pyplot as plt
import os
import emcee
import corner
os.makedirs("SN", exist_ok=True)
```

```python
from astropy.cosmology import FlatLambdaCDM
from astropy.cosmology import WMAP9
from astropy.cosmology import LambdaCDM
from astropy.cosmology import wCDM
from astropy.cosmology import w0waCDM
```

Definimos H0 y temperatura del CMB con valores de WMAP9

```python
H0_units = WMAP9.H(0)
H_0 = H0_units.value
Tcmb0_units = WMAP9.Tcmb(0)
Tcmb_0 = Tcmb0_units.value
```

## Comenzamos con el modelo LCDM con k = 0

```python
cosmo = FlatLambdaCDM(H0 = H_0, Tcmb0 = Tcmb_0, Om0=0.3)
```

```python
def Lum_dist (z):
    return cosmo.luminosity_distance(z).value
```

```python
def mu_datos(z, H, Tcmb):
    dl = Lum_dist(z)
    return 5*np.log10(dl*100000) 
```

```python
data = np.loadtxt('data.dat')
z = data[:,0]
color = data[:,2]
x1 = data[:,3]
mb = data[:,5]
```

```python
mu_i = mu_datos(z, H_0, Tcmb_0)
#mu_i
```

```python
cov_data = np.loadtxt('Pantheon+SH0ES_STATONLY.cov', skiprows = 1)
cov_matrix = cov_data.reshape((1701,1701))
cov_matrix_inv = np.linalg.inv(cov_matrix)
```

### Hasta aquí todo igual al ejercicio original de SN

Definimos las funciones de likelihood y prior igual que en el ejemplo (es importante mantener las dependencias en las variables para el funcionamiento de PyMultiNest.) 

```python
## Copiamos el likelihood de ejercicios anteriores (ahora el vector de parámetros se llama cube para
# que no haya confusión con el ejemplo del tutorial
def loglike(cube, ndim, nparams):
    #alpha, beta, M = cube
    alpha = cube[0]
    beta = cube[1]
    M = cube[2]
    #beta = 3.0
    #alpha = 0.14
    delta = mu_i - (mb+ alpha*x1- beta*color- M)
    chi2 = delta @ cov_matrix_inv @ delta
    return -chi2
```

```python
## Generamos un prior similar al del tutorial
bounds = np.array([[-2., 2.],[-5., 5.],[-21., -15.]])
#bounds = np.array([[-21., -15.]])

def prior(cube, ndim, nparams):
    for i in range(ndim):
        low  = bounds[i,0]
        high = bounds[i,1]
        cube[i] = low + (high-low)*cube[i]    ## el parámetro está entre el valor minimo y maximo de 'bounds'
```

Corremos PyMultiNest, solo necesitamos las funciones de likelihood y prior con las mismas dependencias de parámetros _(cube, ndim, nparams)_. 

```python
parameters = ["a", "b", "M"]
ndim = len(parameters)

pymultinest.run(
    loglike,
    prior,
    ndim,
    outputfiles_basename='SN/',
    resume=False,
    verbose=False
)

json.dump(parameters, open('SN/params.json', 'w'))
```

```python
a = pymultinest.Analyzer(
    n_params=ndim,
    outputfiles_basename='SN/'
)

samples = a.get_equal_weighted_posterior()

corner.corner(
    samples[:,:-1],
    labels=parameters,
    show_titles=True
)
plt.show()
```

## Para un universo con curvatura:

```python
os.makedirs("SN_k", exist_ok=True)     ## Creamos una nueva carpeta para guardar los datos

cosmo_k = LambdaCDM(H0=70, Om0=0.2, Ode0=0.5) 

def Lum_dist (z):
    return cosmo_k.luminosity_distance(z).value

mu_i_k = mu_datos(z, H_0, Tcmb_0)

pymultinest.run(
    loglike,
    prior,
    ndim,
    outputfiles_basename='SN_k/',
    resume=False,
    verbose=False
)

json.dump(parameters, open('SN_k/params.json', 'w'))
```
