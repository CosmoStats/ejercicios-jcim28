```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))


def forward(x, W1, b1, W2, b2):

    z1 = np.dot(W1, x) + b1
    a1 = sigmoid(z1)

    z2 = W2 * a1 + b2
    y_hat = sigmoid(z2)

    return y_hat, a1
```

```python
def compute_cost(y_hat, y):

    eps = 1e-8

    return -(y*np.log(y_hat+eps)
             +(1-y)*np.log(1-y_hat+eps))
```
