### REDES NEURONALES: clasificación binaria

Progreso de las Deep Learning:
Ha mejorado a partir del cambio de la función de activación "Sigmoide" -> "ReLU". Ya que el gradiente de la fucnión sigmoidal tiende a cero en los extremos y los de la función ReLU siempre son 1, esta última permite una convergencia más rápida. Permite hacer los calculos más rápido y hacer pruebas sonbre si las redes funcionan.

"m" denota el número de datos de entrenamientos.

```python
import numpy as np
from PIL import Image
import random
import os
```

```python
cats = []
dogs = []

for archivo in os.listdir('PetImages/Cat'):
    try:
        img = Image.open('PetImages/Cat/' + archivo).convert('RGB')
        cats.append(img.resize((32,32)))
    except:
        pass

for archivo in os.listdir('PetImages/Dog'):
    try:
        img = Image.open('PetImages/Dog/' + archivo).convert('RGB')
        dogs.append(img.resize((32,32)))
    except:
        pass
```

```python
datos = [(x,1) for x in cats] + [(x,0) for x in dogs]     ## labels para gatos y perros

random.shuffle(datos)                                     ## Mezclamos por PARES, conservando los labels
```

```python
X = [x for x,y in datos]
Y = [y for x,y in datos]

X = np.array(X)                                           ## Matriz (m, 32, 32, 3) donde m es el número de datos
Y = np.array(Y)                                            # 32x32 es la dimensión de la imagen y 3 canares (R, G y B)

# Convertir imágenes en vectores
X = X.reshape(len(Y), -1)/255.0                           ## Matriz (m, 32x32x3) es una matriz 2D normalizada
```

```python
corte = int(0.8*len(Y))

X_train = X[:corte]
Y_train = Y[:corte]

X_test = X[corte:]
Y_test = Y[corte:]
```

```python
def sig(z):
    return 1/(1+np.exp(-z))

def dsig(z):
    s = sig(z)
    return s*(1-s)
```

```python
# n_x es el número datos POR imagen, en este caso el número total de pixeles
n_x = X_train.shape[1]
```

```python
# pesos entrada
W1 = np.random.randn(n_x)*0.01
b1 = 0

# pesos oculta 
W2 = np.random.randn()*0.01
b2 = 0
```

```python
alpha = 0.1
```

```python
for epoch in range(100):

    for x,y in zip(X_train,Y_train):

        # Forward

        z1 = np.dot(W1,x) + b1
        a1 = sig(z1)

        z2 = W2*a1 + b2
        a2 = sig(z2)

        # Error

        dz2 = a2 - y

        dW2 = dz2*a1
        db2 = dz2

        dz1 = dz2*W2*dsig(z1)

        dW1 = dz1*x
        db1 = dz1

        # Actualizar

        W2 = W2 - alpha*dW2
        b2 = b2 - alpha*db2

        W1 = W1 - alpha*dW1
        b1 = b1 - alpha*db1
```

```python
correctos = 0

for x,y in zip(X_test,Y_test):

    z1 = np.dot(W1,x) + b1
    a1 = sig(z1)

    z2 = W2*a1 + b2
    a2 = sig(z2)

    pred = 1 if a2 > 0.5 else 0

    if pred == y:
        correctos += 1
```

```python
# Te dice dado una imagen x la probabilidad de que y sea 1, es decir P (y = 1|x)
print("Exactitud:", correctos/len(Y_test)*100, "%")
```

los datos ${x,y}$ se combierten en: $x \rightarrow$ matrices de dimensión $(m,1)$ y $y \rightarrow \{0,1\}$
